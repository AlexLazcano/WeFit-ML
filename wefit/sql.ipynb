```python
import psycopg2
import pandas as pd
```

```python
DBNAME = "test"
HOST = "35.202.86.236"
USER = "alex"
PASSWORD = "alex"
PORT = "5432"

conn = psycopg2.connect(dbname=DBNAME, user=USER, password=PASSWORD, host=HOST, port=PORT)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
    SELECT 
        log_id,
        protein,
        carbohydrates,
        fats,
        user_uuid,
        calories,
        servings,
        log_date,
        food_name,
        image_path,
        serving_size,
        group_id
    FROM food_log
    ORDER BY user_uuid DESC
""")

records = cursor.fetchall()

# Create a DataFrame from the fetched records
df = pd.DataFrame(records, columns=[
    'log_id',
    'protein',
    'carbohydrates',
    'fats',
    'user_uuid',
    'calories',
    'servings',
    'log_date',
    'food_name',
    'image_path',
    'serving_size',
    'group_id'
])

print(df)

```

```python
cursor.execute("""
SELECT
    food_log.user_uuid,
    COUNT(food_log.user_uuid) AS count,
    users.email
FROM
    food_log
JOIN
    users ON food_log.user_uuid = users.user_uuid
GROUP BY
    food_log.user_uuid,
    users.email
ORDER BY
    count DESC
LIMIT 10
""")

records = cursor.fetchall()

# Create a DataFrame from the fetched records
df = pd.DataFrame(records, columns=[
    'user_uuid',
    'count',
    'email'
])

df
```
